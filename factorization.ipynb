```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5,
                        ngram_range=(1, 2), 
                        stop_words='english')
```

```python
from sklearn.feature_selection import chi2
features = tfidf.fit_transform(dcompl['Descriptor']).toarray()
features
```

```python
dcompl['complaint_id'] = dcompl['Complaint Type'].factorize()[0]
complaint_id_df = dcompl[['Complaint Type', 'complaint_id']].drop_duplicates()
```

chi2 selects n features which are having highest value of chi2 statistic

```python
for i,j in k:
    features_chi2 = chi2(features,labels==j)
```

```python
features_chi2    #### Feature vector after applying chi2 test (which returns values for each feature eliminate features not dependent on our class)
```

```python
for i,j in k:
    features_chi2 = chi2(features,labels==j)
    print(features_chi2)
    indices = np.argsort(features_chi2[0])
    print(indices)
    
```

```python
for i,j in k:
    features_chi2 = chi2(features,labels==j)
    print(features_chi2)
    indices = np.argsort(features_chi2[0])
    print(indices)
    feature_names = np.array(tfidf.get_feature_names())[indices]
    print(feature_names)
```

```python
[v for v in feature_names if len(v.split(' ')) == 1]
```
